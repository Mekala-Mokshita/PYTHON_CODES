```python
import numpy as np
import pandas as pd
```

**#creating series**

```python
s=pd.Series([1,3,5,7]) #we can use like ([1,3,5,7,"apple"])
print(s) #displays the series
```

**#COMMON METHODS OF SERIES**

**descriptive stastics:-**

```python
#s.describe():- quick summary of data
s=pd.Series([1,3,5,7,9])
print(s.describe())
```

```python
#s.mean():-computes mean of the data
print(s.mean())
```

```python
#s.std():-computes the standard deviation
print(s.std())
```

```python
#s.min() and s.max():-Comput the minimum and maximum values
print(s.min())
print(s.max())
```

**Data Manipulation**

```python
#s.map(func): Applies a function to each element in the Series.
doubled = s.map(lambda x: x * 2)
print(doubled)
```

```python
#s.apply(func): Similar to map, but more flexible. 
sqrt = s.apply(lambda x: x ** 0.5)
print(sqrt)
```

```python
x = pd.Series([10, 15, 20, 25])

result = x.apply(lambda x: "Even" if x % 2 == 0 else "Odd")
print(result)
```

```python
#s.sort_values(): Sorts the Series.
s=pd.Series([1,3,5,7])
sorted_s = s.sort_values()
print(sorted_s)
```

```python
#s.drop(labels): Drops specified labels from the Series.
dropped=s.drop(1)
print(dropped)
print(s)
```

**#Handling Missing Data**

```python
#s.isnull():-Checks for missing values,returns a series of booleans
print(s.isnull())
```

```python
#s.notnull():Opposite of isnull()
print(s.notnull())
```

```python
#s.fillna(value):-filling missing values
s=pd.Series([1,2,np.nan,4,np.nan])
print(s)
filled=s.fillna(5)
print(filled)
```

```python
#s.dropna():-drops all rows that contain missing values
s_missing=pd.Series([1,2,None,4,5])
dropping_missing=s_missing.dropna()
print(dropping_missing)
```

**#Indexing,Slicing,Filtering**


```python
#s.iloc[]:-integer-location based indexing.(indexing by position)
s=pd.Series([1,3,5,7,9])
print(s.iloc[0])
print(s.iloc[-2])
```

```python
#s.loc[]:-Labeled-based indexing
print(s.loc[0])
print(s.loc[3])
```

```python
s=pd.Series([10,20,30,40],index=['a','b','c','d'])
print(s)
#Accessing using iloc
print(s.iloc[0])
print(s.iloc[1:3])
#accessing using loc
print(s.loc["a"])
print(s.loc["b":"c"])
```

```python
#s[s>n]:-returns greater than n
filtered=s[s>20]
print(filtered)
```

**#Aggregation:-**

```python
#s.sum():sums up the values
print(s.sum())
```

```python
#s.cumsum():cummulative sum
print(s)
print(s.cumsum())
```

```python
#s.aggregate(func):using more than one operations like sum,mean,std
aggregated=s.aggregate(['sum','mean','std'])
print(aggregated)
```

**#merging data frames:-**

```python
role_data={
    'Name':['A','B','C','D'],
    'Role':['founder','manager','it manager','salesman']
}
r_df=pd.DataFrame(role_data)
print(r_df)
contact_data={
    'Name':['A','B','C','D'],
    'phnno':['111-111-1111','222-222-2222','333-333-3333','444-444-4444'],
    'Email':['aa@gmail.com','bb@gmail.com','cc@gmail.com','dd@gmail.com']
    }
c_df=pd.DataFrame(contact_data)
print(c_df)

merged=pd.merge(r_df,c_df,on="Name")
print(merged)
```
